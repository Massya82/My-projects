```python
#import all that we need

import matplotlib.pyplot as plt
import scipy.stats as sts
import pandas as pd
import numpy as np
import math
%matplotlib inline
```

## Покажем центральную предельную теорему на примере непрерывного распределния Рэлея. 
Сгенерируем выборку из 1000 значений случайной величиы

```python
rayleigh_rv = sts.rayleigh(loc=0, scale=1)
sample = rayleigh_rv.rvs(1000)
```

```python
# Зададим ось x
x = np.linspace(0, 5, 1000)

# Теоретическая функция распределения:
cdf = rayleigh_rv.cdf(x)
plt.plot(x, cdf, label='theoretical CDF')

# Эмпирическая функция распределения для 1000 значений
from statsmodels.distributions.empirical_distribution import ECDF
ecdf = ECDF(sample)
plt.step(ecdf.x, ecdf.y, label='ECDF')

plt.ylabel('$f(x)$')
plt.xlabel('$x$')
plt.legend(loc='lower right')
```

Эмпирическая функция и теоретическая практически одинаковы Это происходит из-за большого объема выборки.
Построим гистограмму выборки и график теоретической функции плотности распределения нашей случайной величины

```python
#гистограмма
plt.hist(sample, density=True)

#теоретическая плотность распределения
x = np.linspace(-1, 5, 1000)
pdf = rayleigh_rv.pdf(x)
plt.plot(x, pdf, label='theoretical pdf', alpha=0.5, color='orange')

plt.legend()
plt.ylabel('$f(x)$')
plt.xlabel('$x$')
```

## Применение центральной предельной теоремы
Для каждого значения n - объёма выборки (3, 5, 10, 30, 50) будем генерировать 1000 выборок объёма n, для каждой считать выборочное среднее и строить гистограммы распределений выборочных средних для каждого n, а поверх гистограмм строить графики плотности вероятности для нормальных распределений, которые аппроксимирующт выборочные средние при различных n по ЦПТ

```python
sigma = 1 #по умолчанию

# Теоретические среднее (мат. ожидание) и дисперсия распределния Рэлея
th_mean = math.sqrt(math.pi / 2) * sigma
th_disp = (2 - math.pi / 2)

# Сначала строим гистограмму по выборке объёма n
n = 3
sample_n3 = list()

for i in range(0, 1001):
    tmp_sample = rayleigh_rv.rvs(n)
    cur_mean = sum(tmp_sample) / float(n)
    sample_n3.append(cur_mean)
    
plt.hist(sample_n3, bins=20, density=True)

# Теперь строим график плотности вероятности нормального распределения (по ЦПТ)
norm_rv = sts.norm(th_mean, math.sqrt(th_disp / n))
x = np.linspace(0, 3, 1000)
pdf = norm_rv.pdf(x)
plt.plot(x, pdf, label='current norm pdf', alpha=0.5, color='red')

plt.title('n = 3')
plt.legend(loc="upper right")
plt.ylabel('$f(x)$')
plt.xlabel('$x$')
```
