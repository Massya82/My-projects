<font color='magenta'><b>Комментарий наставника:</b> вот это неправильно. Понимаешь, когда мы заполняем какие-то пропуски мы хотим, чтобы в идеале выполнились 2 условия:
- параметры распределения переменной (среднее и дисперсия) не поменялись;
- зависимости переменной с другими переменными не поменялись.
   

Если у нас ситуация MCAR, то по определению о зависимостях от других переменных мы можем не волноваться. В этом случае, чтобы сохранить среднее, мы либо заполняем средним либо медианой, а вот дисперсией придется пожертвовать. Но в любом случае, каким-то значением из ниоткуда заполнять нельзя - это очень сильно изменит распределение. 

Если у нас не MCAR (т.е. MAR/MNAR), то заполнять без необходимости вообще не стоит. Необходимость может возникнуть, если мы собираемся применить какой-то алгоритм (например машинного обучения), чувствительный к пропускам. Корреляционный анализ к пропускам нечувствителен.

```python
# разобьем наши данные по квантилям, чтобы было проще их анализировать

data['total_income'].quantile([0.25, 0.5, 0.75])
```

```python
#  создадим функцию, которая будет относить людей по группам в соответствии с их доходом

def income_group(total_income):
    if total_income <= 88946.5:
        return 'Низкий доход'
    if total_income <= 135716.0:
        return 'Средний доход'
    if total_income <= 195751.5:
        return 'Высокий доход'
    else:
        return 'Супер доход'
```

```python
data['income_group'] = data['total_income'].apply(income_group)
```

```python
data.head()
```

```python
total_income_test = data.groupby('income_group')['debt'].sum() /\
    data.groupby('income_group')['debt'].count() * 100



```
