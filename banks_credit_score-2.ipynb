# Посмотрим теперь взаимосвязь между уровнем дохода и возвратом кредита

```python
# разобьем наши данные по квантилям, чтобы было проще их анализировать

data['total_income'].quantile([0.25, 0.5, 0.75])
```

```python
#  создадим функцию, которая будет относить людей по группам в соответствии с их доходом

def income_group(total_income):
    if total_income <= 88946.5:
        return 'Низкий доход'
    if total_income <= 135716.0:
        return 'Средний доход'
    if total_income <= 195751.5:
        return 'Высокий доход'
    else:
        return 'Супер доход'
```

```python
data['income_group'] = data['total_income'].apply(income_group)
```

```python
data.head()
```

```python
total_income_test = data.groupby('income_group')['debt'].sum() /\
    data.groupby('income_group')['debt'].count() * 100



```

```python
total_income_test.sort_values()
```

# Посмотрим как разные цели кредита влияют на его возврат 

```python
purpose_keys= []
purpose_set = list(data.purpose.unique())
for value in range(len(purpose_set)):
    if 'образова' in purpose_set[value]:
        purpose_keys.append('Образование')
    elif 'автомо' in purpose_set[value]:
        purpose_keys.append('Автомобиль')
    elif 'свадь' in purpose_set[value]:
        purpose_keys.append('Свадьба')
    elif 'жиль' in purpose_set[value]:
        purpose_keys.append('Недвижимость')
    else:
        purpose_keys.append('Недвижимость')
        
purpose_keys
```

```python
purpose_set
```

```python
purpose_dict = dict(zip(purpose_set, purpose_keys))
data['purpose_rel'] = data['purpose'].map(purpose_dict)

```

```python
pivot_purpose = data.pivot_table(index = 'purpose_rel', columns = ['debt'], values='family_status_id', aggfunc='count')

```

```python


pivot_purpose['relation'] = pivot_purpose[1] / (pivot_purpose[0] + pivot_purpose [1]) *100
```

```python
pivot_purpose.sort_values(by='relation', ascending=True)
```
